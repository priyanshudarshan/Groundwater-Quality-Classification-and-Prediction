```python
gwq2019_WQI = gwq2019.copy()
# Define ideal and standard permissible values for each parameter
parameters = {
    'pH': {'ideal': 7.0, 'standard': 8.5},
    'EC': {'ideal': 0, 'standard': 2000},
    'CO3': {'ideal': 0, 'standard': 9},
    'HCO3': {'ideal': 0, 'standard': 600},
    'Cl': {'ideal': 0, 'standard': 1000},
    'SO4': {'ideal': 0, 'standard': 400},
    'NO3': {'ideal': 0, 'standard': 100},
    'PO4': {'ideal': 0, 'standard': 0.1},
    'TH': {'ideal': 0, 'standard': 600},
    'Ca': {'ideal': 0, 'standard': 200},
    'Mg': {'ideal': 0, 'standard': 100},
    'Na': {'ideal': 0, 'standard': 1756},
    'K': {'ideal': 0, 'standard': 12},
    'F': {'ideal': 0, 'standard': 1.5},
    'TDS': {'ideal': 0, 'standard': 2000},
    'SiO2': {'ideal': 0, 'standard': 50}
}

# Define weights for each parameter as per previous discussion
weights = {
    'pH': 2, 'EC': 1, 'CO3': 1, 'HCO3': 1, 'Cl': 2, 'SO4': 2, 'NO3': 5,
    'PO4': 1, 'TH': 2, 'Ca': 2, 'Mg': 2, 'Na': 1, 'K': 1, 'F': 4, 'TDS': 3, 'SiO2': 1
}

# Normalize weights
total_weight = sum(weights.values())
relative_weights = {param: weight / total_weight for param, weight in weights.items()}

for param, values in parameters.items():
    values['weight'] = relative_weights[param]
print(parameters)

# Classify water quality based on the new WQI ranges
def classify_water_quality(wqi):
    if wqi < 35:
        return 'Excellent'
    elif 35 <= wqi < 45:
        return 'Good'
    elif 45 <= wqi < 55:
        return 'Moderate'
    elif 55 <= wqi < 65:
        return 'Poor'
    elif 65 <= wqi < 75:
        return 'Very Poor'
    else:
        return 'Unsuitable'

# Function to calculate WAWQI
def calculate_wawqi(df):
    # Add a column for each parameter's Qi calculation
    for param, values in parameters.items():
        ideal = values['ideal']
        standard = values['standard']
        weight = values['weight']
        # Calculate Qi and Wi*Qi
        df[f'{param}_Qi'] = 100 * ((df[param] - ideal) / (standard - ideal))
        df[f'{param}_WiQi'] = df[f'{param}_Qi'] * weight

    # Calculate WAWQI for each row
    df['WQI'] = df[[f'{param}_WiQi' for param in parameters]].sum(axis=1) / sum(values['weight'] for values in parameters.values())

    # Classify water quality
    df['Water_Quality'] = df['WQI'].apply(classify_water_quality)
    return df

# Apply the function to calculate WAWQI and classify water quality
gwq2019_WQI = calculate_wawqi(gwq2019_WQI)

# Calculate percentages for each water quality category statewise
statewise_quality = gwq2019_WQI.groupby('State')['Water_Quality'].value_counts(normalize=True).unstack(fill_value=0) * 100

# Calculate overall percentages for the entire dataset
overall_quality = gwq2019_WQI['Water_Quality'].value_counts(normalize=True) * 100

print("Statewise Water Quality Percentage:")
print(statewise_quality)
print("\nOverall Water Quality Percentage for the Country:")
print(overall_quality)
```

```python
gwq2019_WAWQI = gwq2019.copy()
# Define ideal and standard permissible values for each parameter
parameters = {
    'pH': {'ideal': 7.0, 'standard': 8.5},
    'EC': {'ideal': 0, 'standard': 2000},
    'CO3': {'ideal': 0, 'standard': 9},
    'HCO3': {'ideal': 0, 'standard': 600},
    'Cl': {'ideal': 0, 'standard': 1000},
    'SO4': {'ideal': 0, 'standard': 400},
    'NO3': {'ideal': 0, 'standard': 100},
    'PO4': {'ideal': 0, 'standard': 0.1},
    'TH': {'ideal': 0, 'standard': 600},
    'Ca': {'ideal': 0, 'standard': 200},
    'Mg': {'ideal': 0, 'standard': 100},
    'Na': {'ideal': 0, 'standard': 1756},
    'K': {'ideal': 0, 'standard': 12},
    'F': {'ideal': 0, 'standard': 1.5},
    'TDS': {'ideal': 0, 'standard': 2000},
    'SiO2': {'ideal': 0, 'standard': 50}
}

# Calculate proportionality constant K
K = 1 / sum(1 / param['standard'] for param in parameters.values())

# Add the calculated weight for each parameter
for param, values in parameters.items():
    values['weight'] = K / values['standard']
print(parameters)
# Function to classify water quality based on WAWQI
def classify_water_quality(wawqi):
    if wawqi <= 25:
        return 'Excellent'
    elif wawqi <= 50:
        return 'Good'
    elif wawqi <= 75:
        return 'Poor'
    elif wawqi <= 100:
        return 'Very Poor'
    else:
        return 'Unsuitable'

# Function to calculate WAWQI
def calculate_wawqi(df):
    # Add a column for each parameter's Qi calculation
    for param, values in parameters.items():
        ideal = values['ideal']
        standard = values['standard']
        weight = values['weight']
        # Calculate Qi and Wi*Qi
        df[f'{param}_Qi'] = 100 * ((df[param] - ideal) / (standard - ideal))
        df[f'{param}_WiQi'] = df[f'{param}_Qi'] * weight

    # Calculate WAWQI for each row
    df['WAWQI'] = df[[f'{param}_WiQi' for param in parameters]].sum(axis=1) / sum(values['weight'] for values in parameters.values())

    # Classify water quality
    df['Water_Quality'] = df['WAWQI'].apply(classify_water_quality)
    return df

# Apply the function to calculate WAWQI and classify water quality
gwq2019_WAWQI = calculate_wawqi(gwq2019_WAWQI)

# Calculate percentages for each water quality category statewise
statewise_quality = gwq2019_WAWQI.groupby('State')['Water_Quality'].value_counts(normalize=True).unstack(fill_value=0) * 100

# Calculate overall percentages for the entire dataset
overall_quality = gwq2019_WAWQI['Water_Quality'].value_counts(normalize=True) * 100

print("Statewise Water Quality Percentage:")
print(statewise_quality)
print("\nOverall Water Quality Percentage for the Country:")
print(overall_quality)
```

```python
import numpy as np

gwq2019_EWQI = gwq2019.copy()

# Define standard permissible limits for each parameter
parameters = {
    'pH': {'standard': 8.5},
    'EC': {'standard': 2000},
    'CO3': {'standard': 9},
    'HCO3': {'standard': 600},
    'Cl': {'standard': 1000},
    'SO4': {'standard': 400},
    'NO3': {'standard': 100},
    'PO4': {'standard': 0.1},
    'TH': {'standard': 600},
    'Ca': {'standard': 200},
    'Mg': {'standard': 100},
    'Na': {'standard': 1756},
    'K': {'standard': 12},
    'F': {'standard': 1.5},
    'TDS': {'standard': 2000},
    'SiO2': {'standard': 50}
}

# Step 1: Standardize parameters
def standardize_params(df):
    for param in parameters.keys():
        min_val = df[param].min()
        max_val = df[param].max()
        # Standardize to remove unit effects
        df[f'{param}_std'] = (df[param] - min_val) / (max_val - min_val)
    return df

# Step 2: Calculate Entropy for each parameter
def calculate_entropy(df):
    entropy_values = {}
    for param in parameters.keys():
        # Compute Pij for entropy calculation, converting to NumPy array
        Pij = df[f'{param}_std'].astype(np.float64).to_numpy() / df[f'{param}_std'].astype(np.float64).sum()
        # Calculate entropy, handling potential nan values in Pij
        Pij = Pij[np.isfinite(Pij)]  # Remove or handle nan and inf values
        entropy_values[param] = -np.sum(Pij * np.log(Pij + 1e-10)) / np.log(len(df))
    return entropy_values

# Step 3: Calculate entropy weights
def calculate_entropy_weights(entropy_values):
    entropy_weights = {}
    total_entropy = sum(1 - e for e in entropy_values.values())
    for param, entropy in entropy_values.items():
        entropy_weights[param] = (1 - entropy) / total_entropy
    return entropy_weights

# Step 4: Quality Rating and EWQI Calculation
def calculate_ewqi(df, entropy_weights):
    ewqi_values = []
    for index, row in df.iterrows():
        qi_values = []
        for param, values in parameters.items():
            quality_rating = (row[param] / values['standard']) * 100
            weighted_qi = quality_rating * entropy_weights[param]
            qi_values.append(weighted_qi)
        ewqi = sum(qi_values)
        ewqi_values.append(ewqi)
    df['EWQI'] = ewqi_values
    return df

# Step 5: Classify Water Quality
def classify_water_quality(ewqi):
    if ewqi <= 25:
        return 'Excellent'
    elif ewqi <= 50:
        return 'Good'
    elif ewqi <= 100:
        return 'Poor'
    elif ewqi <= 150:
        return 'Very Poor'
    else:
        return 'Unsuitable'

# Full EWQI Calculation Process
gwq2019_EWQI = standardize_params(gwq2019_EWQI)
entropy_vals = calculate_entropy(gwq2019_EWQI)
entropy_weights = calculate_entropy_weights(entropy_vals)
gwq2019_EWQI = calculate_ewqi(gwq2019_EWQI, entropy_weights)
gwq2019_EWQI['Water_Quality'] = gwq2019_EWQI['EWQI'].apply(classify_water_quality)

# Calculate percentages for each water quality category statewise
statewise_quality = gwq2019_EWQI.groupby('State')['Water_Quality'].value_counts(normalize=True).unstack(fill_value=0) * 100

# Calculate overall percentages for the entire dataset
overall_quality = gwq2019_EWQI['Water_Quality'].value_counts(normalize=True) * 100

print("Statewise Water Quality Percentage:")
print(statewise_quality)
print("\nOverall Water Quality Percentage for the Country:")
print(overall_quality)
```

```python
# Create a new DataFrame by copying gwq2019_knn
combined_df = gwq2019.copy()

# Add drinkability columns from each WQI DataFrame
combined_df['WQI_Drinkability'] = gwq2019_wqi['WQI_Drinkability']
combined_df['WAWQI_Drinkability'] = gwq2019_wawqi['WAWQI_Drinkability']
combined_df['EWQI_Drinkability'] = gwq2019_ewqi['EWQI_Drinkability']
combined_df
```

```python
# Create a new column to map water quality categories to drinkability
def map_drinkability(quality):
    """Maps water quality categories to binary drinkability."""
    if quality in ['Unsuitable', 'Poor', 'Very Poor']:
        return '0'  # Not drinkable
    elif quality in ['Excellent', 'Good', 'Medium', 'Moderate']:
        return '1'  # Drinkable
    else:
        return 'Undetermined'  # Handle other cases as needed

# Apply the mapping to each WQI classification column
for wqi_column in ['WQI_Drinkability', 'WAWQI_Drinkability', 'EWQI_Drinkability']:
    combined_df[wqi_column + '_Binary'] = combined_df[wqi_column].apply(map_drinkability)

# Combine drinkability classifications from different WQIs
def determine_drinkability(row):
    """Determine drinkability based on majority vote of WQIs."""
    not_drinkable_count = sum(x == '0' for x in row)
    drinkable_count = sum(x == '1' for x in row)
    if not_drinkable_count >= 2:
        return '0'  # Not drinkable
    elif drinkable_count >= 2:
        return '1'  # Drinkable
    else:
        return 'Undetermined'  # Handle ties or other cases as needed

combined_df['Drinkability'] = combined_df[['WQI_Drinkability_Binary', 'WAWQI_Drinkability_Binary', 'EWQI_Drinkability_Binary']].apply(determine_drinkability, axis=1)

# Calculate the percentage of drinkable water
drinkable_percentage = (combined_df['Drinkability'] == '1').mean() * 100
print(f"Percentage of drinkable water: {drinkable_percentage:.2f}%")
```
